### 정렬

```python
sortKey =['REGIONID','PRODUCT','YEARWEEK']
```

```python
sorted_df = df.sort_values(sortKey)
sorted_df.head()
```

#### 집계

```python
groupKey = ['REGIONID','PRODUCT']#기본키가 regionid,product,yearweek 
```

```python
g_function = ['mean','std','size']
```

```python
group_df = sorted_df.groupby(groupKey)['NEW_QTY'].agg(g_function).reset_index()
```

```python
group_df.head()
```

```python
group_df.rename(columns={'mean':'MEAN','std':'STD','size':'SIZE'},inplace=True)
#컬럼명 담는 괄호..집합표시
```

```python
group_df.head()
```

```python
df.columns
```

```python
group_df.columns
```

#### merge

```python
joinKey = ['REGIONID','PRODUCT']
```

```python
joinData = pd.merge(df,group_df,on=joinKey, how='left')#원래 df에 계산값 붙이니까 innerjoin X
# groupby는 df명.groupby  != merge는 pd.merge
```

```python
joinData.head()
```

## outlier 제거

```python
joinData['REF_QTY']=np.where(joinData.NEW_QTY < (joinData.MEAN - joinData.STD),(joinData.MEAN - joinData.STD),\
                                np.where(joinData.NEW_QTY > (joinData.MEAN + joinData.STD),(joinData.MEAN +
                                                                                            joinData.STD),\
                                        joinData.NEW_QTY))
```

```python
joinData[joinData['REF_QTY']> (joinData.MEAN + joinData.STD)][joinData['NEW_QTY']<(joinData.MEAN + joinData.STD)]
```

```python
sortKey =['REGIONID','PRODUCT','YEARWEEK']
```

```python
sorted_df2 = joinData.sort_values(sortKey)
sorted_df2.head()
```

```python
groupKey = ['REGIONID','PRODUCT']
gfunction2 = ['mean']
groupData2 = sorted_df2.groupby(groupKey)['REF_QTY'].agg(gfunction2).reset_index()
```

```python
groupData2.head()
```

```python
groupData2.rename(columns={'mean':'NEW_MEAN'},inplace=True)
```

```python
joinKey = ['REGIONID','PRODUCT']
joinData2 = pd.merge(sorted_df2,groupData2,on=joinKey, how='left')
```

```python
joinData2.head()
```

### seasonality

```python
recolumns = ['REGIONID','PRODUCT','YEARWEEK','YEAR','WEEK','REF_QTY','NEW_MEAN']
```

```python
middleResult = joinData2[recolumns]
```

```python
middleResult['SEASONALITY'] = np.where(middleResult.mean !=0,\
                                      middleResult.REF_QTY/middleResult.NEW_MEAN, 0)
```

```python
middleResult.head()#연주차별 SEASONALITY
```

#### 주차별 SEASONALITY

```python
finalGroupKey = ['REGIONID','PRODUCT','WEEK']
```

```python
finalResult = middleResult.groupby(finalGroupKey)['SEASONALITY'].agg(['mean']).reset_index()
# agg(['함수']).reset_index()
```

```python
finalResult.rename(columns={'mean':'SEASONALITY'},inplace=True)
```

```python
finalResult.head()
```
