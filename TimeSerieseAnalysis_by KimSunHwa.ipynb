```python
def changDfType(df):
    ipRange = input(str(df.shape[1]) + '개 중 (0부터) "시작, 끝" 숫자 입력 : ').split(',')
    
    ipRange0 = int(ipRange[0])
    ipRange1 = int(ipRange[1])
    
    if ipRange0 == ipRange1:
        colName = df.columns[ipRange0]
        df.loc[:,colName:colName] = df.loc[:,colName:colName].astype(str)
    else:
        start = int(ipRange[0])
        end = int(ipRange[1])+1
        for i in range(start,end):
            colName = df.columns[i]
            df.loc[:,:colName] = df.loc[:,:colName].astype(str)

    return df.dtypes
```

```python
# def dfYearWeekList():
# 범위 입력값
startYearWeek = dfStd.YEARWEEK.min()
endYearWeek = dfStd.YEARWEEK.max()

#df에서 받은 연주차의 년도와 주차 시작/끝 나눠 분리
startY = int(startYearWeek[0:4])
endY = int(endYearWeek[0:4]) +1

startW = int(startYearWeek[4:])
endW = int(endYearWeek[4:])
```

```python
def yearList(sYear,eYear):
    yList=[]
    for year in range(sYear,eYear):
        strYear = str(year)
        yList.append(strYear)
    return yList
```

```python
yList = yearList(startY,endY)
```

```python
def fillDigit(inputList):#두자리 숫자 채워주고, string으로 전환 -> formatting한다면?
    resultList = []
    #monday : month's day
    for inputInt in inputList:
        strDay = str(inputInt)
        strZero = '0'#앞에 0 붙여주기만 되는 함수!
        if len(strDay) == 1:
            resultList.append(strZero + strDay)
        else:
            resultList.append(strDay)
    return resultList
```

```python
def makeYearWeekList(inputYear,startWeek,endWeek):#yearList(startY,endY)실행 후

    wList = []
    for i in range(1,53):
        wList.append(i)#마지막 index = 51
        
    strW_List = fillDigit(wList)#함수
    startW_ListIndex = wList.index(startWeek)
    endW_ListIndex = wList.index(endWeek)
    rList = []
    for i in strW_List[startW_ListIndex:endW_ListIndex+1]:
        rList.append(inputYear+i)
    return rList
```

```python
def dfYearWeekList():#yList,startY,endY,startW,endW 
    resultList = []
    for i in yList:
        if i == str(startY):
            #print('yes'+i)
            resultList.append(makeYearWeekList(i,startW,52))
            
        elif str(startY) < i < str(endY-1):
            #print('mid' + i)
            resultList.append(makeYearWeekList(i,1,52))
   
        else:
            #print('last'+i)
            resultList.append(makeYearWeekList(i,1,endW))
    return resultList
```

```python
yearWeekList = dfYearWeekList()
```
